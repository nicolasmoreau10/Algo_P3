```python
# First, get the count of each surface in the dataframe
values = df_atp['surface'].value_counts()

# Les labels pour le pie chart
labels = df_atp['surface'].value_counts().index

# Création du pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)

# Ajout d'un titre
plt.title("Répartition des terrains/surface")

# Affichage du pie chart
plt.show()
```

###Carpet

```python
# Get the count of each Tourney with Carpet (Plus trop utilisé)
carpet = df_atp['tourney_name'][(df_atp['surface']=='Carpet')].value_counts()

# Defining the data for the bar chart
data = go.Bar(
           x=carpet.index,
           y=carpet.values,
           marker=dict(color='#007EA7')
           )

# Setting the chart titles
layout = go.Layout(
            title="Répartition des tournois sur moquette",
            xaxis=dict(title="Nom du tournois"),
            yaxis=dict(title="Nombre de tournois")
            )
# Creating the bar chart with all arguments
fig = go.Figure(data=[data], layout=layout)

# Displaying the interactive chart
pio.show(fig)
```

## Match

### Evolution du nombres de matchs


```python
new_df = pd.DataFrame({
    'tourney_date': df_atp['tourney_date'],
    'match_num': df_atp['match_num'],
    'surface' : df_atp['surface'],
    'minutes' : df_atp['minutes']
})

# Count the number of matches by year (Use datetime.to_period() method to extract year)
match_year = new_df.groupby(new_df['tourney_date'].dt.to_period('y')).count()['match_num']

match_year.plot(kind='line')
plt.xlabel('Année')
plt.ylabel('Nombre de matchs')
plt.title('Nombre de matchs par année')
plt.show()
```

### Evolution du nombres de matchs/surfaces


```python
# Pivot the DataFrame to create a new DataFrame with columns for each surface and rows for each year
match_year_surface = new_df.pivot_table(index=new_df['tourney_date'].dt.to_period('y'),
                                        columns='surface',
                                        values='match_num',
                                        aggfunc='count')

# Create a line plot for each surface
match_year_surface.plot(kind='line')

# Add x and y axis labels, a title, and a legend
plt.xlabel('Année')
plt.ylabel('Nombre de matchs')
plt.title('Nombre de matchs par année et surface')
plt.legend()

plt.show()
```

## Durées

```python
df_atp['minutes'].describe()
```

```python
new_df['year'] = new_df['tourney_date'].dt.year  # Extract the year from the match date
```

### Durée moyenne / année

```python
# Group the data by year and calculate the mean match time for each year
mean_times_by_year = new_df.groupby('year')['minutes'].mean().reset_index()

# Create the line chart using Plotly Express
fig = px.line(mean_times_by_year, x='year', y='minutes', 
              labels={'year': 'Year', 'minutes': 'Durée du match (en minutes)'},
              title='Durée moyenne des matchs/an')
fig.show()
```

### Durée moyenne / surface

```python
# Group the data by year and surface and calculate the mean match time for each year and surface
mean_times_by_year_surface = new_df.groupby(['year', 'surface'])['minutes'].mean().reset_index()

# Create the line chart using Plotly Express, coloring by surface
fig = px.line(mean_times_by_year_surface, x='year', y='minutes', color='surface',
              labels={'year': 'Année', 'minutes': 'Durée du match (en minutes)', 'surface': 'Terrain'},
              title='Durée moyenne des matchs/an/terrain')
fig.show()
```

```python
df_winner = df_atp.groupby('winner_name')['winner_rank_points'].sum()
df_winner
```

```python
df_loser = df_atp.groupby('loser_name')['loser_rank_points'].sum()
df_loser
```

```python
# Set the figure size
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

# Pandas dataframe
data_box = pd.DataFrame(
    {"Winner Rank Points": df_winner, 
     "Loser Rank Points": df_loser})

# Plot the dataframe
ax = data_box[['Winner Rank Points', 'Loser Rank Points']].plot(kind='box', title='boxplot')

# Display the plot
plt.show()
```

```python
x0 = df_winner
x1 = df_loser

fig = go.Figure()
# Use x instead of y argument for horizontal plot
fig.add_trace(go.Box(x=x0,
                     name='Winner Rank Points'))

fig.add_trace(go.Box(x=x1,
                     name='Loser Rank Points'))

fig.show()
```
